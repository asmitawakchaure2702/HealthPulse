```python
from pinecone.grpc import PineconeGRPC as Pinecone
from pinecone import ServerlessSpec
from langchain.embeddings import HuggingFaceEmbeddings
import itertools

# Initialize Pinecone client
pc = Pinecone(api_key="pcsk_3RQdRy_PAcKXtvR8LbQHdnuNGP3Fy8ck1zTU6q45qLNvpefHGSBdaVKJTn111TGkYoE5XH")

# Connect to the existing Pinecone index
index = pc.Index(host="https://notebook-3ubv4lp.svc.aped-4627-b74a.pinecone.io")  # Replace with your actual index host

# Load embedding model
embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# Function to process text chunks into embeddings and store in Pinecone
def store_text_chunks_in_pinecone(text_chunks, batch_size=1000):
    """Embeds text chunks and inserts them into Pinecone in batches."""
    
    # Generate embeddings for all text chunks
    texts = [chunk.page_content for chunk in text_chunks]  # Extract raw text
    embeddings = embedding_model.embed_documents(texts)  # Get embeddings
    
    # Create vector records for Pinecone
    vector_data = [
        {"id": str(i), "values": embedding, "metadata": {"text": text}}
        for i, (text, embedding) in enumerate(zip(texts, embeddings), start=1)
    ]
    
    # Function to batch process the data
    def batch(iterable, size):
        """Splits data into batches of given size."""
        it = iter(iterable)
        while chunk := list(itertools.islice(it, size)):
            yield chunk
    
    # Batch upsert into Pinecone
    for batch_vectors in batch(vector_data, batch_size):
        index.upsert(vectors=batch_vectors, namespace="text-namespace")

    print(f"✅ Successfully stored {len(text_chunks)} text chunks in Pinecone!")

# Example usage (pass your text_chunks list)
store_text_chunks_in_pinecone(text_chunks)
```
